```python
from OSMPythonTools.nominatim import Nominatim
from OSMPythonTools.overpass import overpassQueryBuilder, Overpass
import pandas as pd
from OSMPythonTools.api import Api
```

```python
nominatim = Nominatim()
areaId = nominatim.query('Poland').areaId()
```

```python

overpass = Overpass()
query1 = overpassQueryBuilder(area=areaId, elementType='node', selector='"brand"="Biedronka"')
query2 = overpassQueryBuilder(area=areaId, elementType='way', selector='"brand"="Biedronka"')
query3 = overpassQueryBuilder(area=areaId, elementType='relation', selector='"brand"="Biedronka"')
result = overpass.query(query1)
result2 = overpass.query(query2)
result3 = overpass.query(query3)
print(result.countElements())
print(result2.countElements())
print(result3.countElements())
```

```python
coordinates = list()
for obj in result._elements:
    coordinates.append((obj._json['lat'], obj._json['lon']))

print(coordinates)
```

```python
for obj in result2._elements:
    api = Api()
    node = api.query(f'node/{obj._json["nodes"][0]}')

    print((node._soup.get('lat')),node._soup.get('lon'))
    coordinates.append((node._soup.get('lat'),node._soup.get('lon')))
```

```python
pd.DataFrame(coordinates).to_csv('cooridinates')
```
